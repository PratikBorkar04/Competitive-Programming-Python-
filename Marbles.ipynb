```python
#marbels

'''I collect marbles (colorful small glass balls) and want to buy boxes to store them.
The boxes come in two types:
Type 1: each such box costs c1 dollars and can hold exactly b1 marbles
Type 2: each such box costs c2 dollars and can hold exactly b2 marbles
I want each box to be filled to its capacity, and also to minimize the total cost of
buying them. Help me find the best way to distribute my marbles among the boxes.'''


t = int(input("Enter total marbles:-"))

c1 = int(input("Enter cost of First Marble:- "))
b1 = int(input("Enter size of First Box:- "))

c2 = int(input("Enter cost of Second Marble:- "))
b2 = int(input("Enter size of Second Box:- "))


if c1 <= c2:
    x1 =  t // b1 
    x2 = x1 * b1

    if x2 == t:
        print("Pass!",x1,0)
    elif x2 != t:
        z1 = x1 - 1
        
        m1 = z1 * b1
        
        m2 = m1 + b2
        
        if t == m2:
            print("pass",z1,1)
        else:
            print("Failed")
    else:
        print("Failed")
        
elif c1 > c2:
    c3 = 0
    c3 = c1
    c1 = c2
    c2 = c3
    
    
    b3 = 0
    b3 = b1
    b1 = b2
    b2 = b3
    
    x1 =  t // b1 
    x2 = x1 * b1

    if x2 == t:
        print("Pass!",x1,0)
    elif x2 != t:
        z1 = x1 - 1
        
        m1 = z1 * b1
        
        m2 = m1 + b2
        
        if t == m2:
            print("pass",z1,1)
        else:
            print("Failed")
    else:
        print("Failed")
'''
Sample Input
43
1 3
2 4

40
5 9
5 12
0

Sample Output
13 1
failed
'''
```
